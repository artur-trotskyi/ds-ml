```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]

# Заміна
df.replace('—', np.nan, inplace=True)

# Перетворення колонок, які не є числовими, в числові
df = df.apply(pd.to_numeric, errors='ignore')

# Виводимо типи колонок після зміни
df.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing_values = df.iloc[:, 1:].isnull().sum()  # Пропускаємо першу колонку

# Перетворюємо в DataFrame і змінюємо назви
missing_values_df = missing_values.reset_index()
missing_values_df.columns = ["Рік", "Кількість пропусків"]

missing_values_df
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці
source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
df.iloc[:-1].tail()  # Видаляє останній рядок
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]

# Перетворення числових стовпців (ігноруючи перший стовпець і останній рядок)
df_numeric = df.iloc[:-1, 1:].apply(pd.to_numeric, errors='coerce')

# Замінюємо на середні значення тільки в числових стовпцях
df_numeric.fillna(df_numeric.mean(), inplace=True)

# Об'єднуємо перший стовпець з числовими значеннями та округлення
df.iloc[:-1, 1:] = df_numeric.round(1)

# Перевірити результат
df
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]

# Виключаємо останній рядок (якщо це середнє значення)
df_cleaned = df.iloc[:-1].copy()

# Конвертуємо 2019 рік у float
df_cleaned['2019'] = pd.to_numeric(df_cleaned['2019'], errors='coerce')

# Обчислюємо середнє значення
mean_2019 = df_cleaned['2019'].mean()

# Вибираємо регіони, де рівень народжуваності вищий за середній
regions_above_avg = df_cleaned[df_cleaned['2019'] > mean_2019]['Регіон'].tolist()

# Виводимо список
regions_above_avg
```

```python
# У якому регіоні була найвища народжуваність у 2014 році?

# Виключаємо останній рядок (якщо це середнє значення)
df_cleaned = df.iloc[:-1].copy()

# Заміна '—' на NaN
df_cleaned.replace('—', np.nan, inplace=True)

# Перетворення стовпця '2014' на числовий тип
df_cleaned['2014'] = pd.to_numeric(df_cleaned['2014'], errors='coerce')

# Тепер сортуємо
df_sorted = df_cleaned.sort_values(by='2014', ascending=False)

# Перевірка результату
max_birth_rate_region = df_sorted[['Регіон', '2014']].iloc[0]
max_birth_rate_region
```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

# Видалення останнього рядка (середнє значення)
df_cleaned = df.iloc[:-1].copy()

# Заміна '—' на NaN
df_cleaned.replace('—', np.nan, inplace=True)

# Перетворення стовпця '2019' на числовий тип
df_cleaned['2019'] = pd.to_numeric(df_cleaned['2019'], errors='coerce')

# Створення стовпчикової діаграми
plt.figure(figsize=(10, 6))
plt.bar(df_cleaned['Регіон'], df_cleaned['2019'], color='skyblue')

# Додавання підписів до осей
plt.xlabel('Регіони')
plt.ylabel('Коефіцієнт народжуваності (2019)')
plt.title('Народжуваність по регіонах України у 2019 році')

# Обертання підписів на осі X для кращого вигляду
plt.xticks(rotation=90)

# Показуємо графік
plt.tight_layout()  # Для того, щоб не було обрізано підписів
plt.show()
```
