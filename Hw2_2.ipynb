```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing_percentage = df.isnull().sum() / len(df) * 100
missing_percentage
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
df_cleaned = df.dropna(axis=1, how='any').join(df['Мова програмування'])

# Виведення перших рядків очищеного DataFrame
df_cleaned.head()
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
missing_percentage = df_cleaned.isnull().sum() / len(df_cleaned) * 100
missing_percentage
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
df_cleaned.dropna(inplace=True)
df_cleaned.head()
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
grouped_by_position = df.groupby('Посада')
position_counts = grouped_by_position.size()
position_counts
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада",
# виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
salary_agg = df.groupby('Посада')['Зарплата на місяць'].agg(['min', 'max'])
salary_agg
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць.
# Використовуйте її для методу apply та створіть новий стовпчик "avg"
file_url = "https://docs.google.com/spreadsheets/d/1nSXJ1oF0GquiLQRqI_I8VOhKNo4B7soGj6nEd5UU1ag/export?format=csv"
df = pd.read_csv(file_url)

def fill_avg_salary(group):
    avg_salary = group['Зарплата на місяць'].mean()
    group['avg'] = avg_salary  # Додаємо середнє значення до кожного рядка групи
    return group

# Групуємо по 'Посада', але виключаємо групування як індекс
df = df.groupby('Посада', group_keys=False, as_index=False).apply(lambda group: fill_avg_salary(group))

# Переглядаємо результат
df[['Посада', 'Зарплата на місяць', 'avg']].iloc[100:106]
```

```python
# Створіть описову статистику за допомогою методу describe для нового стовпчика.
df['avg'].describe()
```
